# 1. 시/도별 평균 건강보혐료

###  30 ~ 59세 평균 건강보험료(원) 데이터
    - 대부분의 직장인을 30~59세대로 추측 
    - Raw Data에서 각 도시별 직장인 평균보험료 추출

```python
# Raw data: 지역별 평균보험료 
insurance_Seoul = pd.read_csv('시군구_서울_인천_경기_강원_별_보험료분위별_현황.csv')
insurance_Busan = pd.read_csv('시군구_부산_대구_울산_경북_경남_별_보험료분위별_현황.csv')
insurance_Daejeon = pd.read_csv('시군구_대전_세종_충북_충남_별_보험료분위별_현황.csv')
insurance_KwangJu = pd.read_csv('시군구_광주_전북_전남_제주_별_보험료분위별_현황.csv')

insurance_Seoul
```

### 각각의 csv  파일 마다 필요한 컬럼만 추출하기

```python
# 서울, 경기, 인천, 강원
avg_ins_Seoul = insurance_Seoul[['시군구별(1)','보험료분위별(1)','2019.4', '2019.3']] 
avg_ins_Seoul = avg_ins_Seoul.drop(index=[0,1,2], axis=0)
avg_ins_Seoul.columns = ['시군구별', '보험료분위별', '평균보험료(원)', '가입자(명)']
avg_ins_Seoul = avg_ins_Seoul.reset_index(drop=True)
avg_ins_Seoul.head()
```

```python
# 부산, 대구, 울산, 경북, 경남
avg_ins_Busan = insurance_Busan[['시군구별(1)','보험료분위별(1)','2019.4', '2019.3']] 
avg_ins_Busan = avg_ins_Busan.drop(index=[0,1,2], axis=0) 
avg_ins_Busan.columns = ['시군구별', '보험료분위별', '평균보험료(원)', '가입자(명)']
avg_ins_Busan = avg_ins_Busan.reset_index(drop=True)
avg_ins_Busan.head()
```

```python
# 대전, 세종, 충북, 충남
avg_ins_Daejeon = insurance_Daejeon[['시군구별(1)','보험료분위별(1)','2019.4', '2019.3']] 
avg_ins_Daejeon = avg_ins_Daejeon.drop(index=[0,1,2], axis=0) 
avg_ins_Daejeon.columns = ['시군구별', '보험료분위별', '평균보험료(원)', '가입자(명)']
avg_ins_Daejeon = avg_ins_Daejeon.reset_index(drop=True)
avg_ins_Daejeon.head()
```

```python
# 평균보험료 = 분위별 세대수 * 평균보험료(원) / 전체 세대수
print(insurance_KwangJu.columns)
```

```python
# 광주, 전북, 전남, 제주
avg_ins_KwangJu = insurance_KwangJu[['시군구별(1)','보험료분위별(1)','2019.4', '2019.3']]
avg_ins_KwangJu = avg_ins_KwangJu.drop(index=[0,1,2], axis=0) 
avg_ins_KwangJu.columns = ['시군구별', '보험료분위별', '평균보험료(원)', '가입자(명)']
avg_ins_KwangJu=avg_ins_KwangJu.reset_index(drop=True)
avg_ins_KwangJu.head()
```

### 각각의 df을 시/도별로 csv 파일로 저장 

```python
avg_ins_Seoul.to_csv("./서울.csv")
avg_ins_Busan.to_csv("./부산.csv")
avg_ins_Daejeon.to_csv("./대전.csv")
avg_ins_KwangJu.to_csv("./광주.csv")
```

### 각각의 시/도의 데이터를 하나의 Df 에 저장: 4개의 데이터를 병합

```python
# 위 4개 데이터 병합
total_Korea_health_insurance = pd.concat([avg_ins_Seoul, avg_ins_Busan, avg_ins_Daejeon, avg_ins_KwangJu], join='outer')
total_Korea_health_insurance=total_Korea_health_insurance.reset_index(drop=True)
total_Korea_health_insurance
total_Korea_health_insurance.to_csv("./전국.csv")
```

```python
avg = []
for i in range(0, len(total_Korea_health_insurance),11):
    avg.append(total_Korea_health_insurance.iloc[i][2])
print(avg)
```

```python
people = []
for i in range(0, len(total_Korea_health_insurance),11):
    people.append(total_Korea_health_insurance.iloc[i][3])
print(people)
```

```python
avg_int = list(map(int, avg)) 
avg_people = list(map(int, people))
```

```python
시군구별 = ['서울','부산','대구','인천','광주','대전','울산','세종','경기','강원','충북','충남','전북','전남','경북','경남','제주']
```

```python
avg_insurance = pd.DataFrame(시군구별, columns =['시군구별'])
avg_insurance.head()
```

# 2. 정신건강검진 결과 

 - 국민건강보험에서 10년단위로 검사를 진행 
 - 30세, 40세,50세가 각각 30대,40대,50대의 정신건강을 대표함 
 - 건강검진을 진행한 30대,40대,50대의 정신건강검사 자료

```python
mental_health = pd.read_csv('전국구 정신건강검사 수검자 최종.csv')
mental_health.drop(columns=['Unnamed: 0'], axis=1) 
mental_health
```

```python
mental_health[['30세 수검자합계']+['40세 수검자합계']+['50세 수검자합계']]
```

```python
Thirty = np.array(mental_health['30세 수검자합계'].tolist())
Forty = np.array(mental_health['40세 수검자합계'].tolist())
Fifty = np.array(mental_health['50세 수검자합계'].tolist())
Thirty_dp = np.array(mental_health['30세우울'].tolist())
Forty_dp = np.array(mental_health['40세우울'].tolist())
Fifty_dp = np.array(mental_health['50세우울'].tolist())
total = Thirty + Forty + Fifty
total_dp = Thirty_dp + Forty_dp + Fifty_dp
```

### 우울증 검사결과가 일상생활에 지장이 있는 경우 우울증환자로 분류됨 

```python
total_mental_health = pd.DataFrame(total, columns =['30~59 수검자 합계'])
total_mental_health['우울증환자'] = total_dp
total_mental_health['시도별'] = mental_health['시도별']
total_mental_health[['시도별', '30~59 수검자 합계', '우울증환자']]
```

```python
divide = total_mental_health['우울증환자']*100000 / avg_insurance['가입자(명)']
adjusted = list(map(float, divide))
```

### 각 도시별 인구수 대비(10만명당) 우울증환자수를 계산 

```python
total_mental_health['우울증환자(10만명당)'] = adjusted
#total_mental_health.to_csv('./우울증환자(10만명당).csv')
```

```python
total_mental_health.drop(['30~59 수검자 합계', '우울증환자'], axis=1)
```

```python
jobless_rate = pd.read_csv('행정구역_시도__연령별_실업률_2019.csv')
jobless_rate.head()
```

```python
jobless_rate = jobless_rate.drop(index = [0, 1, 2, 3], axis=0)
jobless_rate = jobless_rate.drop(columns = ['연령계층별'], axis=0)
jobless_rate = jobless_rate.reset_index(drop=True)
jobless_rate
```

### 각 도시인구수 대비 실업자수를 계산함 (10만명당 실업자수)

```python
avg_rate = []
for i in range(2, len(jobless_rate), 4):
    avg_rate.append(jobless_rate.iloc[i][1]*1000)
print(avg_rate)
```

```python
adjusted_jobless_rate = pd.DataFrame(jobless_rate['시도별'].drop_duplicates())
adjusted_jobless_rate['30~59세 실업자'] = avg_rate
adjusted_jobless_rate.reset_index(drop=True)
```
