```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from shapely.geometry import Point
```

```python
# 시도경계 shapefile
shapefile = 'C:\DataScience\project\geo_data\gadm36_KOR_1.shp'
gdf = gpd.read_file(shapefile)[['NAME_1', 'geometry']]
gdf.columns = ['name','geometry']
gdf
gdf["center"] = gdf["geometry"].centroid
gdf_points = gdf.copy()
gdf_points['city']=['부산', '충북','충남','대구','대전','강원','광주','경기','경북',
                    '경남','인천','제주','전북','전남','세종','서울','울산']
gdf_points.set_geometry("center", inplace = True)
```

```python
ratio_blue = pd.read_csv('연령별 수검자대비 우울증비율 최종.csv', index_col=0)
```

```python
ratio_blue
```

```python
ratio_blue['시도별'] = ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan',
                           'Sejong', 'Gyeonggi-do', 'Gangwon-do', 'Chungcheongbuk-do',
                           'Chungcheongnam-do', 'Jeollabuk-do', 'Jeollanam-do', 'Gyeongsangbuk-do',
                           'Gyeongsangnam-do', 'Jeju']
```

```python
ratio_blue.rename(columns={'시도별':'name'}, inplace=True)
```

```python
ratio_blue
```

```python
#ratio_blue.drop(ratio_blue.columns[4], axis=1, inplace=True)
```

```python
# Geopandas 병합
gdf = pd.merge(gdf, ratio_blue, on='name')
gdf
```

```python
# 시도별 우울비율 지도화 : 30세
f, ax = plt.subplots(1, 1, figsize=(32,16))
gdf.plot(column='30세우울(%)', ax=ax,
                legend=True, legend_kwds={'loc':'lower right'},
                cmap='summer_r', scheme='quantiles')
plt.rc('font', family='Malgun Gothic')
texts = []
for x, y, label in zip(gdf_points.geometry.x, gdf_points.geometry.y, gdf_points['city']):
    texts.append(plt.text(x, y, label, fontsize = 10))
texts
```

```python
# 시도별 우울비율 지도화 : 40세
f, ax = plt.subplots(1, 1, figsize=(32,16))
gdf.plot(column='40세우울(%)', ax=ax,
                legend=True, legend_kwds={'loc':'lower right'},
                cmap='autumn_r', scheme='quantiles')
plt.rc('font', family='Malgun Gothic')
texts = []
for x, y, label in zip(gdf_points.geometry.x, gdf_points.geometry.y, gdf_points['city']):
    texts.append(plt.text(x, y, label, fontsize = 10))
texts
```
