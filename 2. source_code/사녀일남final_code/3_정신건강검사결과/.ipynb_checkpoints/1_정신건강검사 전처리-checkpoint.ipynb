```python
blue1 = pd.read_csv('blue.csv', index_col=0)
```

```python
blue1['30세우울'] = blue1['30중간정도 우울증 의심'] + blue1['30심한 우울증 의심']
blue1['40세우울'] = (blue1['40중간정도 우울증 의심'] + blue1['40심한 우울증 의심'])
blue1['50세우울'] = (blue1['50중간정도 우울증 의심'] + blue1['50심한 우울증 의심'])
```

```python
blue1
```

```python
blue1.drop(blue1.columns[2:4], axis=1, inplace=True)
blue1.drop(blue1.columns[3:5], axis=1, inplace=True)
blue1.drop(blue1.columns[4:6], axis=1, inplace=True)
blue1
```

```python
blue1 = blue1.reindex(columns=['시도별','30세 수검자합계', '30세우울', '40세 수검자합계', '40세우울', '50세 수검자합계','50세우울'])
blue1
```
