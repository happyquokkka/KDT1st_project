```python
blue_korea = pd.read_csv('../data/시·군·구별_우울감_경험률_2019.csv')
blue_korea
```

```python
blue_korea.columns = ['시도별','구별','응답자수(명)', '조율(%)', '조율표준오차', 
                      '표준화율(%)', '표준화율표준오차']
blue_korea.drop([0], inplace=True)
blue_korea
```

# 데이터 속성 설명

#### 표준화율(%)
- 연도 및 지역간 비교를 위해 인구구성 차이에 따른 영향을 표준인구(2005년 추계인구, 통계청)로 보정한 결과

#### 표준화율표준오차
- 표준화율 분포의 표준 오차
    - 표준 오차: 통계의 표본 분포의 표준 편차
