```python
blue = pd.read_csv('../data/전국구정신건강검사_우울증_전처리최종.csv')
```

```python
ppl30 = pd.read_csv('../data/시도별_30세현황데이터_전처리.csv')
ppl40 = pd.read_csv('../data/시도별_40세현황데이터_전처리.csv')
ppl50 = pd.read_csv('../data/시도별_50세현황데이터_전처리.csv')
```

```python
ppl30.drop(ppl30.columns[0], axis=1, inplace=True)
# ppl30
```

```python
ppl40.drop(ppl40.columns[0], axis=1, inplace=True)
ppl50.drop(ppl50.columns[0], axis=1, inplace=True)
```

```python
# 전체 컬럼 중 필요한 컬럼만 추출
blue30 = blue[['시도별', '30세']]
blue40 = blue[['시도별', '40세']]
blue50 = blue[['시도별', '50세']]
```

```python
blue30
```

```python
ratio30 = blue30['30세'] / ppl30['30세'] * 100
ratio30
```

```python
ratio40 = blue40['40세'] / ppl40['40세'] * 100
ratio40
```

```python
ratio50 = blue50['50세'] / ppl50['50세'] * 100
ratio50
```

```python
ratio = pd.DataFrame([ratio30,ratio40,ratio50])
blue_ratio = ratio.T
blue_ratio
```

```python
blue_ratio['시도별'] = blue['시도별']
blue_ratio = blue_ratio.reindex(columns=['시도별','30세','40세','50세'])
```

```python
blue_ratio
```

```python
blue_ratio.to_csv('../data/전국구 총 인구수 대비 우울증결과자.csv')
```
