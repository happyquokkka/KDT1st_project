```python
import pandas as pd
import numpy as np
```

```python
insurance_Seoul = pd.read_csv('시군구_서울_인천_경기_강원_별_보험료분위별_현황.csv')
```

```python
avg_ins_Seoul = insurance_Seoul[['시군구별(1)','보험료분위별(1)','2019','2019.1', '2019.2']] 
# 필요없는 인덱스 삭제 (row 삭제)
avg_ins_Seoul = avg_ins_Seoul.drop(index=[0,1,2], axis=0) 
# 컬럼이름 변경
avg_ins_Seoul.columns = ['시군구별', '보험료분위별', '세대수(세대)', '평균보험료(원)', '적용인구(명)']
# 인덱스 번호 수정
avg_ins_Seoul = avg_ins_Seoul.reset_index(drop=True)
avg_ins_Seoul
```

```python
insurance_Busan = pd.read_csv('시군구_부산_대구_울산_경북_경남_별_보험료분위별_현황.csv')
```

```python
avg_ins_Busan = insurance_Busan[['시군구별(1)','보험료분위별(1)','2019','2019.1', '2019.2']] 
# 필요없는 인덱스 삭제 (row 삭제)
avg_ins_Busan = avg_ins_Busan.drop(index=[0,1,2], axis=0) 
# 컬럼이름 변경
avg_ins_Busan.columns = ['시군구별', '보험료분위별', '세대수(세대)', '평균보험료(원)', '적용인구(명)']
# 인덱스 번호 수정
avg_ins_Busan = avg_ins_Busan.reset_index(drop=True)
avg_ins_Busan
```

```python
insurance_Daejeon = pd.read_csv('시군구_대전_세종_충북_충남_별_보험료분위별_현황.csv')
```

```python
avg_ins_Daejeon = insurance_Daejeon[['시군구별(1)','보험료분위별(1)','2019','2019.1', '2019.2']] 
# 필요없는 인덱스 삭제 (row 삭제)
avg_ins_Daejeon = avg_ins_Daejeon.drop(index=[0,1,2], axis=0) 
# 컬럼이름 변경
avg_ins_Daejeon.columns = ['시군구별', '보험료분위별', '세대수(세대)', '평균보험료(원)', '적용인구(명)']
# 인덱스 번호 수정
avg_ins_Daejeon = avg_ins_Daejeon.reset_index(drop=True)
avg_ins_Daejeon
```

```python
insurance_KwangJu = pd.read_csv('시군구_광주_전북_전남_제주_별_보험료분위별_현황.csv')
```

```python
# 평균보험료 = 분위별 세대수 * 평균보험료(원) / 전체 세대수
print(insurance_KwangJu.columns)
#['시군구별(1)','보험료분위별(1)','2019.1', '2019.2'] 데이터 추출
avg_ins_KwangJu = insurance_KwangJu[['시군구별(1)','보험료분위별(1)','2019','2019.1', '2019.2']] 
# 필요없는 인덱스 삭제 (row 삭제)
avg_ins_KwangJu = avg_ins_KwangJu.drop(index=[0,1,2], axis=0) 
# 컬럼이름 변경
avg_ins_KwangJu.columns = ['시군구별', '보험료분위별', '세대수(세대)', '평균보험료(원)', '적용인구(명)']
# 인덱스 번호 수정
avg_ins_KwangJu=avg_ins_KwangJu.reset_index(drop=True)
avg_ins_KwangJu
# avg_ins_KwangJu.set_index('시군구별')
# avg_ins_KwangJu
```

```python
total_Korea_health_insurance = pd.concat([avg_ins_Seoul, avg_ins_Busan, avg_ins_Daejeon, avg_ins_KwangJu], join='outer')
# 인덱스 번호 수정
total_Korea_health_insurance=total_Korea_health_insurance.reset_index(drop=True)
total_Korea_health_insurance
```
